```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
fig, ax = plt.subplots(figsize=(10, 5))

# Define vessel and valve regions
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 2)

# Draw blood vessel (laminar flow)
vessel_x = np.linspace(-1, 6, 100)
vessel_y_upper = np.full_like(vessel_x, 1.5)
vessel_y_lower = np.full_like(vessel_x, -1.5)

ax.fill_between(vessel_x, vessel_y_upper, vessel_y_lower, color="lightblue", alpha=0.3, label="Blood Vessel")

# Add laminar flow arrows in vessel
for i in np.linspace(-0.5, 2, 5):
    ax.arrow(i, 0.8, 0.5, 0, head_width=0.15, head_length=0.2, fc="blue", ec="blue")
    ax.arrow(i, -0.8, 0.5, 0, head_width=0.15, head_length=0.2, fc="blue", ec="blue")

# Draw heart valve
valve_x = [3, 3.2, 3.5, 3.8, 4]
valve_y_upper = [1.5, 1.2, 1, 1.2, 1.5]
valve_y_lower = [-1.5, -1.2, -1, -1.2, -1.5]

ax.plot(valve_x, valve_y_upper, color="black", linewidth=3, label="Heart Valve")
ax.plot(valve_x, valve_y_lower, color="black", linewidth=3)

# Add turbulent flow arrows in valve region
for i in np.linspace(3.1, 3.9, 4):
    ax.arrow(i, 0.4, 0.2, -0.2, head_width=0.15, head_length=0.15, fc="red", ec="red")
    ax.arrow(i, -0.4, 0.2, 0.2, head_width=0.15, head_length=0.15, fc="red", ec="red")

# Labels
ax.text(0, 1.7, "Laminar Flow (Vessel)", fontsize=12, color="blue")
ax.text(3.2, 1.7, "Turbulent Flow (Valve)", fontsize=12, color="red")
ax.text(3.2, -1.7, "Regurgitant Jet", fontsize=12, color="darkred")

# Hide axes
ax.axis("off")

# Title
ax.set_title("Blood Flow Characteristics in Vessels vs. Heart Valves", fontsize=14)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Define vessel and valve regions
ax.axvline(x=0.3, color='black', linestyle='--', label="Valve Region")
ax.axvline(x=0.7, color='black', linestyle='--')

# Define parabolic velocity profile for vessel flow
x_vessel = np.linspace(0, 0.3, 100)
y_vessel = 2 * (1 - (x_vessel / 0.3) ** 2)
ax.plot(x_vessel, y_vessel, 'b', label=r'Vessel Flow: $w = 2 \nu \left( 1 - \left( \frac{r}{d/2} \right)^2 \right)$')

# Indicate turbulent and regurgitant flow in valve
x_valve = np.linspace(0.3, 0.7, 100)
y_turbulent = np.sin(10 * x_valve) * 0.5 + 0.5  # Random turbulence representation
ax.plot(x_valve, y_turbulent, 'r', linestyle='dotted', label="Turbulent Flow in Valve")

# Annotate key regions
ax.text(0.15, 1.5, "Laminar Flow in Vessel", fontsize=10, color='blue')
ax.text(0.5, 1.2, "Turbulent Jet in Stenotic Valve", fontsize=10, color='red')
ax.text(0.5, 0.3, "Regurgitant Backflow", fontsize=10, color='purple')

# Set labels and title
ax.set_title("Blood Flow in Vessels vs. Valves", fontsize=12)
ax.set_xlabel("Radial Position (r)", fontsize=10)
ax.set_ylabel("Velocity (w)", fontsize=10)
ax.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vessel diameter and velocity for parabolic profile
d = 1  # Normalized vessel diameter
nu = 1  # Normalized maximum velocity at the center

# Define radial positions
r = np.linspace(-d/2, d/2, 100)
w = 2 * nu * (1 - (r / (d/2))**2)  # Parabolic velocity profile

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Plot parabolic velocity profile (laminar flow in vessels)
ax.plot(w, r, label=r'$w = 2 \nu \left( 1 - \left( \frac{r}{d/2} \right)^2 \right)$', color='b', linewidth=2)
ax.fill_betweenx(r, 0, w, color='blue', alpha=0.2)

# Add arrows to indicate flow direction in vessels (laminar)
for i in np.linspace(-d/2, d/2, 7):
    ax.arrow(0.5, i, 0.3, 0, head_width=0.05, head_length=0.1, fc='blue', ec='blue')

# Add turbulent flow representation for valves (chaotic flow)
for i in np.linspace(-d/2, d/2, 5):
    ax.arrow(1.5, i, np.random.uniform(-0.3, 0.3), np.random.uniform(-0.3, 0.3),
             head_width=0.05, head_length=0.05, fc='red', ec='red')

# Labels for different flow types
ax.text(0.4, 0.6, "Laminar Flow in Vessels", fontsize=12, color='blue', weight='bold')
ax.text(1.4, 0.6, "Turbulent Flow in Valves", fontsize=12, color='red', weight='bold')
ax.text(0.6, -0.8, r"Parabolic Flow: $w = 2 \nu \left( 1 - \left( \frac{r}{d/2} \right)^2 \right)$", fontsize=10, color='blue')
ax.text(1.6, -0.8, "Irregular Jet Flow & Regurgitant Backflow", fontsize=10, color='red')

# Transition arrow between vessel and valve flow
ax.annotate("", xy=(1.2, 0), xytext=(0.8, 0), arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.9, 0.1, "Flow through Valve", fontsize=12, color='black')

# Axis limits and labels
ax.set_xlim([-0.2, 2])
ax.set_ylim([-0.7, 0.7])
ax.set_xlabel("Flow Velocity", fontsize=12)
ax.set_ylabel("Radial Position", fontsize=12)
ax.set_title("Comparison of Laminar Vessel Flow and Turbulent Valve Flow", fontsize=14, weight='bold')

# Hide Y-axis ticks
ax.set_yticks([])

# Show plot
plt.show()
```
